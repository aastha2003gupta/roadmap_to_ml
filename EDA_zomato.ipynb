```python
df_country=pd.read_excel('/content/Country-Code[1].xlsx')
```

```python
df_country.head()
```

```python
#combining country codes with both datasets
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head(2)
```

```python
#check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
```

```python
#check top 3 countries
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Observation: Zomato max records from india , followed by usa , uk

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings.head()
```

Observation : 4.5-4.9:excellent
4.0-4.4: very good
3.5-3.9: good
3.0-3.4:average
2.5-2.9:average
2.0-2.4:poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

Observation: not rated count very high , max ratings are between 2.5-3.4

```python
#count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```
