MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

add lables

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine values')
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x,np.sin(x),'g.')
plt.plot(x,np.cos(x),'r+')

plt.show()
```

bar plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
languages=['English','French','Spanish','latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('people')
plt.show()
```

pie chart

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
languages=['English','French','Spanish','latin','German']
people=[100,50,150,40,70]
ax.pie(people,labels=languages,autopct='%1.lf%%')
plt.show()
```

Scatter plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

3D SCATTER PLOT

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```

Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

build in datasets of seaborn

```python
#total bill vs tip
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#setting theme for plots
sns.set_theme()
```

```python
#visualize the data
sns.relplot(data=tips,x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
iris=sns.load_dataset('iris')
iris.head()
```
