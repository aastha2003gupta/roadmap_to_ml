```python
numeric_features = list(vg_data.select_dtypes(include=["float64", "int64"]).columns)
numeric_features
```

Distribution of numeric features

```python
fig, axes = plt.subplots(nrows=len(numeric_features), ncols=1, figsize=(8, 6 * len(numeric_features)))

# Plot histograms for each numeric feature
for idx, feature in enumerate(numeric_features):
    ax = axes[idx]
    vg_data[feature].plot(kind='hist', ax=ax, bins=20)
    ax.set_title(f'Histogram of {feature}')
    ax.set_xlabel(feature)

plt.tight_layout()
plt.show()




```
